```
# Importing libraries

import glob
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
from keras.utils.np_utils import to_categorical
```

```
# Loading data from Google Drive

from google.colab import drive
drive.mount('/content/drive')
```

```
# Storing Data in DataFrame and Labeling the Data for Training 
# df1 for Category 1 Data
df1 = pd.DataFrame()

cv_img = []
for img in glob.glob("/content/drive/My Drive/Dataset/train/category 1/*.png"):
    n= cv2.imread(img,cv2.IMREAD_GRAYSCALE)
    cv_img.append(n)

df1['image'] = cv_img

# df2 for Category 2 Data
df2 = pd.DataFrame()

cv_img = []
for img in glob.glob("/content/drive/My Drive/Dataset/train/category 2/*.png"):
    n= cv2.imread(img,cv2.IMREAD_GRAYSCALE)
    cv_img.append(n)

df2['image'] = cv_img

# df3 for Category 3 Data
df3 = pd.DataFrame()

cv_img = []
for img in glob.glob("/content/drive/My Drive/Dataset/train/category 3/*.png"):
    n= cv2.imread(img,cv2.IMREAD_GRAYSCALE)
    cv_img.append(n)

df3['image'] = cv_img

# df4 for Category 4 Data
df4 = pd.DataFrame()

cv_img = []
for img in glob.glob("/content/drive/My Drive/Dataset/train/category 4/*.png"):
    n= cv2.imread(img,cv2.IMREAD_GRAYSCALE)
    cv_img.append(n)

df4['image'] = cv_img
```

```


# Labeling data for training 

df1['label'] = 0
df2['label'] = 1
df3['label'] = 2
df4['label'] = 3
```

```
# Combining Data in single DataFrame 

df = pd.DataFrame()
df = pd.concat([df1,df2,df3,df4],sort=False,ignore_index=True)
```

```
# Loading Testing Data 

TEST_DIR = []
for img in glob.glob("/content/drive/My Drive/Dataset/test/*.png"):
    n= cv2.imread(img)
    TEST_DIR.append(n)
    
    
```

```
# Preparing  training data 

X = np.array(df['image'])
y = np.array(df['label'])
```

```
# Building neural networks model

X = X/255.0

model = Sequential()
model.add(Conv2D(64,(3,3),input_shape = (800,800,1) ))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))

model.add(Dense(4))
model.add(Activation('sigmoid'))

```

```
# Compiling model

model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
```

```
# Categorically encode labels

y = to_categorical(y)
```

```
# Training model

model.fit(X, y,epochs= 3, batch_size = 10 )
```
